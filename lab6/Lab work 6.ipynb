### Завдання:

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()
```

Створити візуалізацію, аналогічно рисунку 
 - перші 3 графіки візуалізувати, викорстовуючи значення із df
 - останній це пряма пропорційність

```python
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
df['a'].plot()

plt.subplot(2, 2, 2)
df['b'].plot()

plt.subplot(2, 2, 3)
df['c'].plot()


plt.subplot(2, 2, 4)
plt.plot(df.index)

plt.tight_layout()
plt.show()
```

Створити візуалізацію, аналогічно рисунку 

```python
plt.figure(figsize=(15, 15))
plt.subplot(2, 1, 1)
df['a'].plot()
plt.show()
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`

```python
plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
df['a'].plot(kind='line')

plt.subplot(2, 1, 2)
df['b'].plot(kind='line')
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 1, 1)
df['a'].plot()

plt.subplot(2, 1, 1)
df['b'].plot()
plt.show()
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`
- задайте стиль 'darkgrid' за допомогою команди `sns.set_style`

```python
plt.figure(figsize=(10, 10))
sns.set_style('darkgrid')

plt.subplot(2, 1, 1)
df['a'].plot()

plt.subplot(2, 1, 1)
df['b'].plot()
plt.show()
```

Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

- для колонки `c` використайте жовтий колір та товщину `1` і маркер `o`


```python
plt.figure(figsize=(10, 10))

sns.set_style('darkgrid')

plt.subplot(2, 1, 1)
df['a'].plot(color='red', linestyle='-.')

plt.subplot(2, 1, 1)
df['b'].plot(color='orange', linewidth=10)

plt.subplot(2, 1, 1)
df['c'].plot(color='yellow', marker='o', linewidth=1)
plt.show()
```

Створити візуалізацію, аналогічно рисунку 
- і не забудьте про легенду :)

```python
plt.figure(figsize=(9, 5))

sns.set_style('darkgrid')

plt.subplot(3, 1, 1)
plt.plot(df['a'], label='Line a')
plt.legend(loc='center right')

plt.subplot(3, 1, 2)
plt.plot(df['b'], label='Line b')
plt.legend(loc='center left')

plt.subplot(3, 1, 3)
plt.plot(df['c'])

plt.tight_layout()
plt.show()
```

Створити візуалізацію, аналогічно рисунку 
- використайте томатний колір та відстань між стовбцями 0.5

```python
plt.figure(figsize=(10, 5))

plt.bar(range(len(df['a'])), df['a'], label='a', color='#FF6347', alpha=0.5)
plt.legend(loc='upper right')

```

Створити візуалізацію, аналогічно рисунку 
- добавте всі підписи та правильний маркер

```python
plt.figure(figsize=(20, 8))
plt.plot(df["a"], marker="^")
plt.xticks(df.index)
plt.title("This is a Title", fontsize=18)
plt.xlabel("This is X Axis", fontsize=24)
plt.ylabel("Column 'a' values", fontsize=12)

plt.tight_layout()
plt.show()
```
